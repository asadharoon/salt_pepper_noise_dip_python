```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import imread
import math
from skimage.util import random_noise
from skimage.filters import rank
from scipy import ndimage
```

# Original Image

```python
original_image=imread("circuit_image.png",False,'L')
```

```python
plt.imshow(original_image,plt.cm.gray)
```

```python
processed_image=original_image.copy()
```

# Adding Pepper Noise

```python
noise_img = random_noise(processed_image, mode='s&p')
noise_image=noise_img*255
```

```python
plt.imshow(noise_image,plt.cm.gray)
```

```python
hist=plt.hist(noise_image.flatten(),bins=256,range=[0,255])
```
